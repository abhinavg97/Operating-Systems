```python
import matplotlib.pyplot as plt
import numpy as np
plt.title('K vs Average time')

plt.gca().set_color_cycle(['red','blue'])
plt.plot([1000, 2000, 3000, 4000, 5000], [0.000050, 0.000025, 0.000033, 0.000037, 0.000160], label = 'RW')
plt.plot([1000, 2000, 3000, 4000, 5000], [0.000100, 0.000175, 0.000083, 0.000112, 0.000080], label = 'RW_Fair')
plt.legend(loc = 'best')
plt.xlabel('K')
plt.ylabel('Averafe time taken (in us)')
# plt.show()
plt.savefig('good.jpg')
```
